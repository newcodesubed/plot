```python
import numpy as np
from scipy.optimize import root
import matplotlib.pyplot as plt

def f(x):
    return x**3 - x**2 -2*x + 1

def df(x):
    return 3*x**2 - 2*x - 2

def newton_raphson(initial_guess, tolerance=1e-6, max_iterations=100):
    x = initial_guess
    iterations = 0
    roots = []
    
    while True:
        x_next = x - f(x) / df(x)
        iterations += 1
        
        if abs(x_next - x) < tolerance or iterations >= max_iterations:
            break
        
        x = x_next
    
    if iterations >= max_iterations:
        print("Maximum number of iterations reached.")
    else:
        roots.append(x_next)
    
    return roots

# Define initial guess and parameters for Newton-Raphson method
initial_guess = np.linspace(-10, 10, 20)  # Initial guesses
tolerance = 1e-6
max_iterations = 100

# Find roots using Newton-Raphson method
newton_roots = []
for guess in initial_guess:
    newton_roots.extend(newton_raphson(guess, tolerance, max_iterations))

# Find roots using scipy.optimize.root()
scipy_result = root(f, initial_guess)
scipy_roots = scipy_result.x

# Print the roots obtained from both methods
print("Roots obtained from Newton-Raphson method:", newton_roots)
print("Roots obtained from scipy.optimize.root():", scipy_roots)

# Plot the function
x_values = np.linspace(-5, 5, 400)
y_values = f(x_values)
plt.figure(figsize=(8, 6))
plt.plot(x_values, y_values, label='$f(x) = x^3 - x^2 -2x + 1$')
plt.axhline(0, color='black', linewidth=0.5)
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Plot of $f(x) = x^3 - x^2 -2x + 1$')
plt.grid(True)
plt.legend()

# Plot roots found by Newton-Raphson method
plt.scatter(newton_roots, [0]*len(newton_roots), color='red', label='Roots (Newton-Raphson)')
# Plot roots found by scipy.optimize.root()
plt.scatter(scipy_roots, [0]*len(scipy_roots), color='blue', label='Roots (scipy.optimize.root())')
plt.legend()

# Show plot
plt.show()
```
