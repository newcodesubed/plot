```python
import numpy as np
import matplotlib.pyplot as plt

# Define the function
def f1(x):
    return x / (x**2 + 1)

# Define the range of x values
x_values = np.linspace(0, 5, 400)

# Calculate the corresponding y values
y_values1 = f1(x_values)

# Plot the function
plt.figure(figsize=(8, 6))
plt.plot(x_values, y_values1, label='$f(x) = \\frac{x}{x^2 + 1}$', color='blue')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Plot of $y = \\frac{x}{x^2 + 1}$')
plt.grid(True)
plt.legend()
plt.show()
```

```python
# Define the function
def f2(x):
    return np.exp(x)

# Calculate the corresponding y values
y_values2 = f2(x_values)

# Plot the function
plt.figure(figsize=(8, 6))
plt.plot(x_values, y_values2, label='$f(x) = e^x$', color='green')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Plot of $y = e^x$')
plt.grid(True)
plt.legend()
plt.show()
```
