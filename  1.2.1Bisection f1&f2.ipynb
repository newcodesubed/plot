```python
import numpy as np
import matplotlib.pyplot as plt

# Define the function
def f(x):
    return x**2 - x - 1

# Bisection method implementation
def bisection(a, b, e, Nmax):
    # Initialize lists to store iterations
    iterations = []
    roots = []

    condition = True
    N = 1
    
    while condition and N <= Nmax:
        g = f(a)
        x = (a + b) / 2
        if f(x) < 0:
            a = x
        else:
            b = x
        iterations.append(N)
        roots.append(x)
        N += 1
        condition = abs(f(x) - g) > e
        
    if N > Nmax:
        print("Maximum number of iterations reached. Consider changing initial values or tolerance.")
    else:
        print("Required root is:", x)
        
    return iterations, roots

# Define initial values and parameters
a = 1
b = 2
e = 0.0001
Nmax = 10

# Run the bisection method
iterations, roots = bisection(a, b, e, Nmax)

# Plot the function
x_values = np.linspace(0, 3, 100)
plt.plot(x_values, f(x_values), label='f(x)')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Plot of the function f(x)')
plt.grid(True)
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)

# Plot the iterations
plt.scatter(roots, [f(x) for x in roots], color='red', label='Iterations')
for i, (x, fx) in enumerate(zip(roots, [f(x) for x in roots])):
    plt.text(x, fx, f'({iterations[i]})', fontsize=8, verticalalignment='bottom')

plt.legend()
plt.show()
```
