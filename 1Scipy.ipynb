```python
import numpy as np
from scipy.optimize import root
import matplotlib.pyplot as plt

# Define the function
def f(x):
    return x**2 - x - 1

# Bisection method implementation
def bisection(a, b, e):
    roots = []
    condition = True
    
    while condition:
        g = f(a)
        x = (a + b) / 2
        if f(x) < 0:
            a = x
        else:
            b = x
        condition = abs(f(x) - g) > e
        roots.append(x)
        
    return roots

# Define initial interval and tolerance for bisection method
a = 1
b = 2
e = 0.0001

# Find roots using bisection method
bisection_roots = bisection(a, b, e)

# Find roots using scipy.optimize.root()
optimize_root_result = root(f, [a, b])

# Extract roots from the result
scipy_roots = optimize_root_result.x

# Print the roots obtained from both methods
print("Roots obtained from bisection method:", bisection_roots)
print("Roots obtained from scipy.optimize.root():", scipy_roots)

# Plot the function
x_values = np.linspace(-2, 3, 400)
y_values = f(x_values)
plt.figure(figsize=(8, 6))
plt.plot(x_values, y_values, label='$f(x) = x^2 - x - 1$')
plt.axhline(0, color='black', linewidth=0.5)
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Plot of $f(x) = x^2 - x - 1$')
plt.grid(True)
plt.legend()

# Plot roots found by bisection method
plt.scatter(bisection_roots, [0]*len(bisection_roots), color='red', label='Roots (Bisection Method)')
# Plot roots found by scipy.optimize.root()
plt.scatter(scipy_roots, [0]*len(scipy_roots), color='blue', label='Roots (scipy.optimize.root())')
plt.legend()

# Show plot
plt.show()
```
